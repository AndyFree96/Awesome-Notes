### 调整线条的颜色与风格

要修改线条的颜色，可以通过color参数，它支持各种颜色值的字符串。

```python
plt.plot(x, np.sin(x - 0), color = "blue") # 标准颜色名称
plt.plot(x, np.sin(x - 1), color = "g") # 缩写颜色代码（rgbcmyk）
plt.plot(x, np.sin(x - 2), color = "0.75") # 范围在0～1的灰度值
plt.plot(x, np.sin(x - 3), color = "#FFDD44") # 十六进制（RRGGBB, 00~FF）
plt.plot(x, np.sin(x - 4), color = "chartreuse") # HTML颜色名称
plt.plot(x, np.sin(x - 5), color = (1.0, 0.2, 0.3)) # RGB元组，范围在0~1
```

类似的，可以通过linestyle参数改变线条风格：

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

```python
plt.plot(x, x + 0, linestyle='-') # 实线
plt.plot(x, x + 1, linestyle='--') # 虚线
plt.plot(x, x + 2, linestyle='-.') # 点划线
plt.plot(x, x + 3, linestyle=':') # 点线
```

也可以将linestyle和color组合在一起：

```python
plt.plot(x, x + 0, '-g') # 绿色实线
plt.plot(x, x + 1, '--c') # 青色虚线
plt.plot(x, x + 2, '-.k') # 黑色点划线
plt.plot(x, x + 3, ':r') # 红色点线
```

详情请看文档：https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

### 调整坐标轴上下限

Matplotlib会自动选择合适的坐标轴上下限，我们也可以自己手动设置，使用plt.xlim()和plt.ylim()即可。

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

也可以逆序设置坐标轴刻度：

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

使用plt.axis()函数传入`[xmin, xmax, ymin, ymax]`对应的值，可以让我们用一行代码设置x轴和y轴的上下限。

```python
plt.plot(x, np.cos(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.cos(x - 1))
plt.axis([11, -1, 1.5, -1.5])
```

通过plt.axis()函数也可以让图形内容自动收紧坐标轴，不留空白区域。

```python
plt.plot(x, np.cos(x - 1))
plt.axis([11, -1, 1.5, -1.5]);
```

```python
plt.plot(x, np.cos(x - 1))
plt.axis([11, -1, 1.5, -1.5])
plt.axis('tight')
```

更多关于plt.axis的内容，请看文档：https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.axis.html

### 设置图形标签

这一部分我们将学习设置标签，包括图形标题、坐标轴标题以及图例。

```python
plt.plot(x, np.sin(x))
plt.title("A sine curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

当在单个ax绘制多条线时，我们可以给每条线设置图例。

```python
plt.plot(x, np.sin(x), "-g", label="sin(x)")
plt.plot(x, np.cos(x), ":b", label="cos(x)")
plt.axis("equal")
plt.legend();
```

在面向对象接口画图时，我们可以使用ax.set()方法一次性设置所有属性。

```python
ax = plt.axes()
ax.plot(x ,np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2,2),xlabel="x", ylabel="sin(x)", title="A sine curve");
```

## 简易散点图

散点图由单个的点、圆圈或其他形状构成。

### 用plt.plot绘制散点图

```python
x = np.linspace(0, 10, 30)
```

```python
plt.plot(x, np.sin(x), "o", color="black");
```

```python
plt.plot(x, np.sin(x), 'o');
```

函数的第三个参数是一个字符串，表示图形符号的类型。

```python
rng = np.random.RandomState(0)

for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8)
```

还可以与线条、颜色代码组合起来，画出一条连接散点的线。

```python
plt.plot(x, np.sin(x), '-ok'); # 直线（-）、圆圈（o）、黑色（k）
```

plt.plot支持许多设置线条和散点属性的参数。

```python
plt.plot(x, np.sin(x), '-p', color='gray',
         markersize=15, linewidth=4, markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

### 用plt.scatter绘制散点图

```python
plt.scatter(x, np.sin(x), marker='o');
```

```python
plt.scatter(x, np.sin(x));
```

plt.scatter和plt.plot的区别在于，前者在创建散点图时具有更高的灵活性，可以单独控制每个散点与数据匹配，也可以让每个散点具有不同的属性。

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar(); # 显示颜色条
```

颜色自动映射成颜色条，散点的大小以像素为单位。这样，散点的颜色和大小就可以在可视化图中显示多维数据的信息。

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
features = iris.data.T
```

```python
features.shape
```

```python
iris.target
```

```python
plt.scatter(features[0],features[1], alpha=0.2,
            s=100 * features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

上图中的每个点的坐标值`(x,y)`分别表示花萼的长度与宽度，而点的大小表示花瓣的宽度，三种颜色对应三种不同类型的鸢尾花。

### plt.plot与plot.scatter的效率对比

当数据量小的时候，两者在效率上差异不大。但当数据量变大到几千个散点时，plt.plot的效率将大大高于plt.scatter。这是因为plt.scatter会对每个散点进行单独的大小与颜色的渲染，因此渲染器会消耗更多的资源。而plt.plot中，散点基本都彼此复制，因此整个数据集中的所有点的颜色、尺寸只需配置一次。

### 连续误差

```python
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
model = lambda x : x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)
```

```python
gp = GaussianProcessRegressor(random_state=100)
```

```python
gp.fit(xdata[:, np.newaxis], ydata)
```

```python
xfit = np.linspace(0, 10, 1000)
yfit= gp.predict(xfit[:, np.newaxis])
```

## 密度图与等高线图

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
Z = f(X, Y)
```

```python
Z
```

```python
plt.style.use('seaborn-white')
```

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```
